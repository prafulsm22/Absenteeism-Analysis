```python
import pandas as pd 
import numpy as np 
```

```python
raw_data = pd.read_csv('Absenteeism_data.csv')
raw_data

df = raw_data.copy()
raw_data.head()
```

```python
raw_data.info()
```

```python
data = raw_data.drop(['ID'],axis=1)
data.describe(include='all')
```

### Group the reason for absence 

```python
sorted(pd.unique(data['Reason for Absence']))
```

### number 20 is missing

```python
reasons_columns = pd.get_dummies(data['Reason for Absence'])
reasons_columns
```

```python
reasons_columns['check'] = reasons_columns.sum(axis = 1)
reasons_columns['check'].sum(axis=0)
```

```python
reasons_columns['check'].unique()
```

```python
reasons_columns = reasons_columns.drop(['check'], axis = 1)
```

```python
reason_columns = pd.get_dummies(data['Reason for Absence'], drop_first = True)
reason_columns
```

```python
reason_type1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
data = data.drop(['Reason for Absence'],axis=1)
data.head()
```

## CONCATENATE THE COLUMNS

```python
data = pd.concat([data, reason_type1, reason_type2 , reason_type3 , reason_type4], axis=1)
data
```

### Assign column names in a more meaningful way

```python
data.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1','Reason_2', 'Reason_3', 'Reason_4']
```

```python
data.columns = column_names
data.head()
```

```python
column_names_ordered = ['Reason_1','Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
        'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data = data[column_names_ordered]
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format = '%d/%m/%Y')
data.head()
```

### Extract the month value ( 1 - 12)

```python
list_months = []

for i in range(data.shape[0]):
    list_months.append(data['Date'][i].month)
    
len(list_months)
```

```python
data['Month Value'] = list_months
data.head()
```

### Extract the days of the week ( 0 - 6)

```python
  def date_to_weekday(date_value):
        return date_value.weekday()
```

```python
data['Day of the week'] = data['Date'].apply(date_to_weekday)
data.head()
```

```python
data.columns.values
```

```python
column_names_ordered = ['Reason_1','Reason_2', 'Reason_3', 'Reason_4', 'Month Value',
       'Day of the week', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
        'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data = data[column_names_ordered]
data = data.drop(['Date'],axis=1)
data.head()
```
